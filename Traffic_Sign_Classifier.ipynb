### Model Architecture

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

def LeNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    W1 = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean=mu, stddev=sigma))
    b1 = tf.Variable(tf.zeros(6))
    conv_layer1 = tf.nn.conv2d(x, W1, strides=[1,1,1,1], padding='VALID') + b1

    # Activation.
    conv_layer1 = tf.nn.relu(conv_layer1, name='relu1')

    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv_layer1 = tf.nn.max_pool(conv_layer1, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')

    # Layer 2: Convolutional. Output = 10x10x16.
    W2 = tf.Variable(tf.truncated_normal((5, 5, 6, 16), mean=mu, stddev=sigma))
    b2 = tf.Variable(tf.zeros(16))
    conv_layer2 = tf.nn.conv2d(conv_layer1, W2, strides=[1,1,1,1], padding='VALID') + b2
    
    # Activation.
    conv_layer2 = tf.nn.relu(conv_layer2, name='relu2')

    # Pooling. Input = 10x10x16. Output = 5x5x16.
    conv_layer2 = tf.nn.max_pool(conv_layer2, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')    

    # Flatten. Input = 5x5x16. Output = 400.
    fl = tf.contrib.layers.flatten(conv_layer2)

    # Layer 3: Fully Connected. Input = 400. Output = 120.
    W3 = tf.Variable(tf.truncated_normal(shape=(400, 120), mean=mu, stddev=sigma))
    b3 = tf.Variable(tf.zeros(120))
    layer3 = tf.matmul(fl, W3) + b3
    
    # Activation.
    layer3 = tf.nn.relu(layer3)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    W4 = tf.Variable(tf.truncated_normal(shape=(120, 84), mean=mu, stddev=sigma))
    b4 = tf.Variable(tf.zeros(84))
    layer4 = tf.matmul(layer3, W4) + b4
    
    # Activation.
    layer4 = tf.nn.relu(layer4)

    # Layer 5: Fully Connected. Input = 84. Output = 43.
    W5 = tf.Variable(tf.truncated_normal(shape=(84, 43), mean=mu, stddev=sigma))
    b5 = tf.Variable(tf.zeros(43))
    logits = tf.matmul(layer4, W5) + b5
    
    return logits

````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
# Preparation

EPOCHS = 10
BATCH_SIZE = 256

x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
````

````python
# Operations

rate = 0.007

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
# Train

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train_scaled)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train_scaled, y_train = shuffle(X_train_scaled, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train_scaled[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate(X_valid_scaled, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './traffic_sign')
    print("Model saved")
````

````python
# Evaluate the model with training set
with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('.'))

    test_accuracy = evaluate(X_train_scaled, y_train)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````

````python
# Evaluate the model with test set

# Convert test set to gray scale and normalize it, just like training and validation sets.
X_test_gray = np.array([np.dot(x[...,:3], [0.299, 0.587, 0.114]).reshape(32, 32, 1) for x in X_test])
X_test_scaled = np.array([(x.astype(float)-128.0)/128.0 for x in X_test_gray])

with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('.'))

    test_accuracy = evaluate(X_test_scaled, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````

### Load and Output the Images

````python
# The GTSRB training images are downloaded and unzipped to project directory beforehand.

# Randomly select 5 image classes.
import csv
import os
import random
from PIL import Image
from datetime import datetime
random.seed(datetime.now())

thedir = "GTSRB/training"
dirs = [name for name in os.listdir(thedir) if os.path.isdir(os.path.join(thedir, name))]
dirs = random.sample(dirs, 5)
labels = [int(d) for d in dirs]

# From each 5 classes, randomly select 1 image.
# Crop ROI of the images to increase prediction accuracy.
images = []
for c in dirs:
    # Randomly select one image from a class.
    image_name = random.choice([i for i in os.listdir(os.path.join(thedir, c)) if os.path.splitext(i)[1]=='.ppm'])
    image_file = os.path.join(thedir, os.path.join(c, image_name))
    image = Image.open(image_file)
    image.load()
    
    # Get the ROI of the image.
    ROI = [] # [x1, y1, x2, y2]
    with open(os.path.join(thedir, os.path.join(c,'GT-{}.csv'.format(c))), 'r') as f:
        file_meta = csv.reader(f, delimiter=';')
        for row in file_meta:
            if image_name == row[0]:
                ROI = [int(i) for i in row[3:7]]
                break
    
    cropped = image.crop(ROI)
    images.append(cropped)

# Load signnames.csv
name_label_map = {}
with open('signnames.csv', 'r') as f:
    signnames = csv.reader(f)
    
    for row in signnames:
        name_label_map[row[0]] = row[1]
    
# Show images.
%matplotlib inline

for i in range(0, len(images)):
    plt.figure()
    plt.subplot(321+i)
    plt.title(name_label_map[str(labels[i])])
    plt.imshow(images[i])
    
````
